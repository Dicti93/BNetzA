# **Dataset**

The dataset comprises 194 network providers and 904 variables, which are broadly categorized into three key operational dimensions: **Capacity (Kapazität)**, **Service (Dienstleistung)**, and **Transport (Transport)**.

To streamline the analysis, these variables have been grouped into "Technical Blocks”, which are engineered combinations of related parameters representing the core aspects of these dimensions. The Technical Blocks serve as a framework for grouping similar variables according to their operational focus:

**Capacity Blocks (K1-K10):** These blocks focus on parameters related to peak loads and installed capacities, considering both energy intake and output across different network levels.

**Service Blocks (DA1-DA8, DG1-DG5):** These blocks are concerned with network extension (e.g., cable lengths) and granularity (e.g., metering points), which reflect how networks are structured and how they serve customers.

**Transport Blocks (T1-T4):** These blocks track energy flows across network levels, distinguishing between aggregated and disaggregated data according to operational needs.

**2. Feature Analysis**

**- Sparsity:** As shown in Figure 3, a substantial portion of the data (314 out of 894 features) contains over 90% zero values, indicating a high degree of sparsity. This could introduce noise into the model, so careful feature selection will be necessary to ensure model accuracy. We will address this issue during the data preprocessing phase.

**- Skewness vs. Proportion of Zeros:** The relationship between skewness and the proportion of zero values indicates that higher proportions of zeros are associated with greater skewness (see Figure 4). This issue will need to be addressed during preprocessing, potentially through transformations or other advanced techniques.



```python
cols_to_drop = ['BNR_NNR', 'NameOrg', 'NameShort', 'dmu', 'dDateData', 'cTOTEXn_RP3', 'cTOTEXs_RP3', 'cTOTEXs',  'BNR','dmuName']
df = df.drop(columns=cols_to_drop)

df.describe()
```

```python
# Calculate the proportion of zeros in each column
zero_proportion = (df == 0).mean()

# Define bins for the proportions (10% increments)
bins = np.arange(0, 1.1, 0.1)  # Adjusted to include 1.0 explicitly

# Count the number of columns in each bin
zero_counts = np.histogram(zero_proportion, bins=bins)[0]

# Create labels for the x-axis (e.g., '0%', '10%', ..., '100%')
labels = [f'{int(x*100)}%' for x in bins[:-1]]

# Modify the label for 90% to ">90%"
labels[-1] = '>90%'

# Plot the distribution
plt.figure(figsize=(12, 6))
bars = plt.bar(labels, zero_counts, color='coral')

# Add text labels on top of each bar
for bar, count in zip(bars, zero_counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             str(count), ha='center', va='bottom')

plt.title('Distribution of Columns by Proportion of Zero Values')
plt.xlabel('Proportion of Zeros')
plt.ylabel('Number of Columns')
plt.show()
```

**3. Correlation and Multicollinearity**

**- Heatmap:** Figure 5 displays the top 10 aggregated-level features most correlated with cTOTEXn reveals extremely high correlations (above 0.95), especially between energy-related features like yInjection.tot and yEnergy.delivered.tot, which have a correlation of 0.98. The heatmap also shows that these features have high correlations with each other, suggesting potential multicollinearity, where variables might be measuring similar aspects of the data. Hierarchical clustering further groups these features, indicating closely related constructs. During model building, we will carefully evaluate and address this potential issue.



```python
# Create a list of columns to keep: those with '.tot' and 'xTOTEXn'
columns_to_keep = [col for col in df.columns if '.tot' in col] + ['cTOTEXn']

# Subset the DataFrame to keep only the desired columns
df_filtered = df[columns_to_keep]

# Display the first few rows of the filtered DataFrame (for verification)
print(df_filtered.shape)
```

```python
# Calculate correlation with 'cTOTEXn'
correlation = df_filtered.corr()['cTOTEXn'].drop('cTOTEXn')

# Sort correlations from high to low
sorted_correlation = correlation.sort_values(ascending=False)

print(sorted_correlation)
```

```python
# Select top 10 correlations
top_n = 10
top_correlations = sorted_correlation.head(top_n)
```

```python
# Add 'cTOTEXn' to the list of top correlated features
top_features = df_filtered[top_correlations.index.tolist() + ['cTOTEXn']]

# Calculate the correlation matrix
corr_matrix = top_features.corr()

# Plot the heatmap with clustering
sns.clustermap(corr_matrix, cmap='coolwarm', annot=True, linewidths=0.5, figsize=(10, 10))
plt.title('Heatmap of Top 10 Correlated Features Including cTOTEXn')
plt.show()
```
