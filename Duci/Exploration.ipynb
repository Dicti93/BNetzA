Drop cTOTEXn_RP3 and cTOTEXs_RP3 since the other TOTEX variables do not have any missing values.                                  

```python
df_2 = df.drop(['BNR_NNR', 'NameOrg', "NameShort", "dmu", "dDateData", "cTOTEXn_RP3", "cTOTEXs_RP3", "cTOTEXs", "dmuName", "BNR"], axis=1)
df_2 = df_2.fillna(0)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


# Specify the target variable
target_column = 'cTOTEXn'

# Define X and y
X = df_2.drop(columns=[target_column])
y = df_2[target_column]

# Initialize a DataFrame to store feature importances
feature_importances_df = pd.DataFrame(index=X.columns)

# Run the Random Forest 10 times
for i in range(10):
    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
    
    # Initialize the Random Forest model
    rf = RandomForestRegressor(n_estimators=100, random_state=i)
    
    # Train the model
    rf.fit(X_train, y_train)
    
    # Get feature importances
    feature_importances = rf.feature_importances_
    
    # Add the importances to the DataFrame
    feature_importances_df[f'Run_{i+1}'] = feature_importances

# Calculate the mean and standard deviation of the importances across the runs
feature_importances_df['Mean_Importance'] = feature_importances_df.mean(axis=1)
feature_importances_df['Std_Deviation'] = feature_importances_df.std(axis=1)

# Sort by mean importance
feature_importances_df = feature_importances_df.sort_values(by='Mean_Importance', ascending=False)

# Display the feature importances
print(feature_importances_df)

# Plot the feature importances
plt.figure(figsize=(12, 8))
plt.bar(feature_importances_df.index, feature_importances_df['Mean_Importance'], yerr=feature_importances_df['Std_Deviation'])
plt.title('Feature Importances across 10 Random Forest Runs')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt

# Specify the target variable
target_column = 'cTOTEXn'

# Define X and y
X = df_2.drop(columns=[target_column])
y = df_2[target_column]

# Initialize a DataFrame to store feature importances and errors
feature_importances_df = pd.DataFrame(index=X.columns)
errors_list = []

# Run the Random Forest 10 times
for i in range(10):
    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
    
    # Initialize the Random Forest model
    rf = RandomForestRegressor(n_estimators=100, random_state=i)
    
    # Train the model
    rf.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = rf.predict(X_test)
    
    # Calculate MSE, RMSE, and MAPE
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    
    # Append the errors to the list
    errors_list.append({'Run': i+1, 'MSE': mse, 'RMSE': rmse, 'MAPE': mape})
    
    # Get feature importances
    feature_importances = rf.feature_importances_
    
    # Add the importances to the DataFrame
    feature_importances_df[f'Run_{i+1}'] = feature_importances

# Create a DataFrame from the errors list
errors_df = pd.DataFrame(errors_list)

# Calculate the mean and standard deviation of the importances across the runs
feature_importances_df['Mean_Importance'] = feature_importances_df.mean(axis=1)
feature_importances_df['Std_Deviation'] = feature_importances_df.std(axis=1)

# Sort by mean importance
feature_importances_df = feature_importances_df.sort_values(by='Mean_Importance', ascending=False)

# Display the feature importances
print(feature_importances_df)

# Display the MSE, RMSE, and MAPE
print(errors_df)

# Plot the feature importances
plt.figure(figsize=(12, 8))
plt.bar(feature_importances_df.index, feature_importances_df['Mean_Importance'], yerr=feature_importances_df['Std_Deviation'])
plt.title('Feature Importances across 10 Random Forest Runs')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=90)
plt.show()

# Plot MSE, RMSE, and MAPE for each run
plt.figure(figsize=(12, 6))
plt.plot(errors_df['Run'], errors_df['MSE'], marker='o', label='MSE')
plt.plot(errors_df['Run'], errors_df['RMSE'], marker='o', label='RMSE')
plt.plot(errors_df['Run'], errors_df['MAPE'], marker='o', label='MAPE')
plt.title('MSE, RMSE, and MAPE across 10 Random Forest Runs')
plt.xlabel('Run')
plt.ylabel('Error')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
import matplotlib.pyplot as plt

# Specify the target variable
target_column = 'cTOTEXn'

# Define X and y
X = df_2.drop(columns=[target_column])
y = df_2[target_column]


# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Apply PCA
# Choose the number of components to explain 95% of the variance
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Function to evaluate and print model performance
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    print(f"{model.__class__.__name__} - MSE: {mse}, RMSE: {rmse}, MAPE: {mape}")

# Models
models = [
    Ridge(alpha=1.0),
    Lasso(alpha=0.01),
    ElasticNet(alpha=0.01, l1_ratio=0.5),
    SVR(kernel='linear', C=1.0),
    GradientBoostingRegressor(n_estimators=100, learning_rate=0.1),
    RandomForestRegressor(n_estimators=100, random_state=42)
]

# Evaluate each model
for model in models:
    evaluate_model(model, X_train, X_test, y_train, y_test)

# Feature importances for GradientBoostingRegressor and RandomForestRegressor
def plot_feature_importances(model, X_pca, model_name):
    model.fit(X_train, y_train)
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        indices = np.argsort(importances)[::-1]
        plt.figure(figsize=(12, 8))
        plt.title(f"Feature Importances - {model_name}")
        plt.bar(range(X_pca.shape[1]), importances[indices], align="center")
        plt.xlabel("Principal Components")
        plt.ylabel("Importance")
        plt.show()

# Plot feature importances for applicable models
for model in [GradientBoostingRegressor(n_estimators=100, learning_rate=0.1), RandomForestRegressor(n_estimators=100, random_state=42)]:
    plot_feature_importances(model, X_pca, model.__class__.__name__)
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


# Specify the target variable
target_column = 'cTOTEXn'

# Define X and y
X = df_2.drop(columns=[target_column])
y = df_2[target_column]

# Step 1: Preprocess the Data
# Handle missing values (if any)
X = X.fillna(X.mean())

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Apply PCA
# Choose the number of components to explain 95% of the variance
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# Step 3: Calculate the importance of each original feature
pca_components = pca.components_
explained_variance = pca.explained_variance_ratio_

# Feature importance in the original feature space
original_feature_importance = np.dot(pca_components.T, explained_variance)

# Create a DataFrame to display the importances
feature_importance_df = pd.DataFrame(original_feature_importance, index=df_2.columns[:-1], columns=['Importance'])
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Display the feature importances
print(feature_importance_df)

# Plot the feature importances in the original feature space
plt.figure(figsize=(12, 8))
plt.bar(feature_importance_df.index, feature_importance_df['Importance'])
plt.title('Feature Importances in Original Feature Space')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=90)
plt.show()
```
